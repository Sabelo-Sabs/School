```python
from numpy import linalg as LA
```

```python
# 3.1 Define the above matrices in Python.
M1 = np.array([[1,1,-2],[3,2,4],[0,3,0]])
M2 = np.array([[1,2,2],[2,2,0],[0,-1,1]])
M3 = np.array([1,3,5,6])
```

```python
M1
```

```python
M2
```

```python
M3
```

```python
# 3.2 Product of M1 and M2 and store it in K.
K = M1.dot(M2)
K
```

```python
# 3.3 Determinant of K.
LA.det(K)
```

```python
# 3.4 Eigen‐values and eigen‐vectors of K.
eig_val ,eig_vec = LA.eig(K)
```

```python
eig_val
```

```python
eig_vec
```

```python
# 3.5 Find the inverse of K and store it in a new matrix M4.
M4 = LA.inv(K)
M4
```

```python
# 3.6 The transpose of M4.
np.transpose(M4)
```
